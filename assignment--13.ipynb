```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv(r'C:\Users\saite\Downloads\Day_13_Pharma_data.csv')  # Replace with the correct file path

# Step 1: Data cleaning
# Check for missing values
print(df.isnull().sum())

# Fill or drop missing values (if necessary)
df = df.dropna()  # You can also fill missing values using df.fillna(value)

# Check for duplicates
print(df.duplicated().sum())
df = df.drop_duplicates()  # Drop duplicates if any

# Step 2: Visualizations

# 1. Bar plot showing total sales per region
plt.figure(figsize=(10, 6))
sales_per_region = df.groupby('Region')['Sales'].sum()
sales_per_region.plot(kind='bar', color='skyblue')
plt.title('Total Sales per Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2. Scatter plot to visualize the relationship between Marketing_Spend and Sales
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Marketing_Spend', y='Sales', color='green')
plt.title('Relationship between Marketing Spend and Sales')
plt.xlabel('Marketing Spend')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()

# 3. Boxplot comparing drug effectiveness across different age groups
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Age_Group', y='Effectiveness', palette='Set2')
plt.title('Drug Effectiveness across Different Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Effectiveness')
plt.tight_layout()
plt.show()

# 4. Line plot showing the sales trend for each product over different trial periods
# If the 'Product' column doesn't exist or is named differently, replace 'Product' with the correct column name
# Example: If 'Product' doesn't exist, use 'Region' or any other column suitable for the line plot
# Uncomment below if 'Product' column exists and is named correctly
# sns.lineplot(data=df, x='Trial_Period', y='Sales', hue='Product', marker='o')

# If 'Product' doesn't exist, remove the hue argument (or replace 'Product' with an actual column name)
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Trial_Period', y='Sales', marker='o')
plt.title('Sales Trend over Different Trial Periods')
plt.xlabel('Trial Period')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()

# 5. Heatmap of the correlation between Sales, Marketing_Spend, and Effectiveness
correlation_data = df[['Sales', 'Marketing_Spend', 'Effectiveness']]
correlation_matrix = correlation_data.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation between Sales, Marketing Spend, and Effectiveness')
plt.tight_layout()
plt.show()

# Step 3: Summarize any patterns or trends observed
# Example analysis:
# - Positive correlation between marketing spend and sales suggests that higher marketing investment leads to increased sales.
# - Drug effectiveness seems to vary across different age groups, with certain age groups showing higher effectiveness.
# - Regional sales distribution shows which regions are performing better.
# - Sales trends over trial periods can help identify peak performance times and potential seasonal patterns.
```
